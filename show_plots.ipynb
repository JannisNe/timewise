```python
from pymongo import MongoClient
from IPython.display import SVG, display
from ampel.util.compression import decompress
import cairosvg
```

```python

client = MongoClient("mongodb://localhost:27017")
db = client["test_ampel_masked"]
```

```python

def svg_to_pdf(svg_str, filename="output.pdf"):
    cairosvg.svg2pdf(bytestring=svg_str, write_to=filename)
```

```python

i = 39
doc = db["t1"].find_one({"stock": i})
svg_string = decompress(doc["meta"][0]["plot"]["svg"])

display(SVG(svg_string))
svg_to_pdf(svg_string, f"{i}_diagnostic.pdf")
```
